```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
status = ('Divorced', 'Single', 'Married', 'Widowed')
y_pos = np.arange(len(status))
numbers = [400, 556, 678, 123]
plt.bar(y_pos, numbers, align='center', alpha =0.5)
plt.ylabel('Count')
plt.title('Marital Status of Narnians')
plt.show()
```

```python
pokemon = pd.read_csv("/Users/kieranmcavoy/Documents/Data Science/Big Data Technologies/Lab 2/pokemon_alopez247.csv")
```

```python
pokemon['Type_1'].value_counts().plot.bar()
plt.ylabel('count')
plt.title('Pokemon types')
plt.show()
```

```python
pokemon['Type_2'].value_counts().nlargest(3).sort_values(ascending=False).plot.bar()
plt.title('Top 5 pokemon types in dataset')
plt.show()
```

```python
pokemon['Attack'].value_counts().nlargest(10).sort_values(ascending = True).plot.bar()
plt.title('Pokemon Attack stats')
plt.ylabel('Frequency Used')
plt.xlabel('Power Level')
plt.show()
```
